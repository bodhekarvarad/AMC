```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

np.random.seed(42)

zones = ["Zone_1", "Zone_2", "Zone_3", "Zone_4"]

def generate_sensor_data(timestamp):

    zone = np.random.choice(zones)

    flow = np.random.normal(100, 5)
    pressure = np.random.normal(55, 3)
    acoustic = np.random.normal(25, 2)

    anomaly = 0

    # Leak condition
    if np.random.rand() < 0.05:
        pressure *= np.random.uniform(0.5, 0.7)
        flow *= np.random.uniform(1.2, 1.5)
        acoustic *= np.random.uniform(1.5, 2.0)
        anomaly = 1

    return [timestamp, zone, flow, pressure, acoustic, anomaly]


# Generate 10,000 timestamps (1-second interval)
start_time = datetime.now()

data = [
    generate_sensor_data(start_time + timedelta(seconds=i))
    for i in range(10000)
]

df = pd.DataFrame(data, columns=[
    "timestamp", "zone", "flow", "pressure", "acoustic", "anomaly"
])

df.shape
df.head()

df.to_csv("pipeline_sensor_data_10000.csv", index=False)
```

```python

df = pd.read_csv("pipeline_sensor_data_10000.csv")
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp")
```

```python
df = pd.get_dummies(df, columns=["zone"])
```

```python
df["flow_ma"] = df["flow"].rolling(window=10).mean()

df["pressure_drop_rate"] = df["pressure"].diff()

df["acoustic_change"] = df["acoustic"].diff()

df["flow_pressure_ratio"] = df["flow"] / df["pressure"]
```

```python
df.bfill(inplace=True)
```

```python
X = df.drop(columns=["timestamp", "anomaly"])
y = df["anomaly"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    n_estimators=200,
    contamination=0.05,
    random_state=42
)

model.fit(X_scaled)
```

```python
df["anomaly_score"] = model.decision_function(X_scaled)

df["predicted_anomaly"] = model.predict(X_scaled)

df["predicted_anomaly"] = df["predicted_anomaly"].map({1: 0, -1: 1})
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y, df["predicted_anomaly"]))
print(classification_report(y, df["predicted_anomaly"]))
```

```python
import joblib

joblib.dump(model, "anomaly_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))
plt.plot(df["pressure"], label="Pressure")

plt.scatter(
    df.index[df["predicted_anomaly"] == 1],
    df["pressure"][df["predicted_anomaly"] == 1],
    color="red",
    label="Leak Detected"
)

plt.legend()
plt.show()
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import plotly.express as px
from collections import deque
import folium
from streamlit_folium import st_folium
import time

# ---------------- PAGE CONFIG ----------------
st.set_page_config(page_title="Smart Water Monitoring", layout="wide")

st.markdown("<h1 style='text-align: center; color: #0A75AD;'>🚰 Smart Water Pipeline Monitoring Dashboard</h1>", unsafe_allow_html=True)
st.markdown("---")

# ---------------- LOAD MODEL ----------------
model = joblib.load("anomaly_model.pkl")
scaler = joblib.load("scaler.pkl")

# ---------------- SESSION STATE ----------------
if "run" not in st.session_state:
    st.session_state.run = False

if "flow" not in st.session_state:
    st.session_state.flow = deque(maxlen=50)
    st.session_state.pressure = deque(maxlen=50)
    st.session_state.acoustic = deque(maxlen=50)

if "map_key" not in st.session_state:
    st.session_state.map_key = 0

# ---------------- CONTROL BUTTONS ----------------
c1, c2 = st.columns(2)

if c1.button("▶️ Start Monitoring"):
    st.session_state.run = True

if c2.button("⏹ Stop Monitoring"):
    st.session_state.run = False

# ---------------- ZONE SETUP ----------------
zones = ["Zone_A", "Zone_B", "Zone_C", "Zone_D"]

zone_map_model = {
    "Zone_A": "zone_Zone_1",
    "Zone_B": "zone_Zone_2",
    "Zone_C": "zone_Zone_3",
    "Zone_D": "zone_Zone_4"
}

zone_coordinates = {
    "Zone_A": [18.5204, 73.8567],
    "Zone_B": [18.5314, 73.8446],
    "Zone_C": [18.5420, 73.8650],
    "Zone_D": [18.5100, 73.8700]
}

NORMAL_FLOW = 100
NORMAL_PRESSURE = 55

model_features = [
    'flow', 'pressure', 'acoustic',
    'zone_Zone_1', 'zone_Zone_2', 'zone_Zone_3', 'zone_Zone_4',
    'flow_ma', 'pressure_drop_rate', 'acoustic_change', 'flow_pressure_ratio'
]

# ---------------- FUNCTIONS ----------------
def get_severity(flow, pressure):
    score = (NORMAL_FLOW - flow) + (NORMAL_PRESSURE - pressure)

    if score < 5:
        return "🟢 LOW", "green"
    elif score < 15:
        return "🟡 MEDIUM", "orange"
    else:
        return "🔴 HIGH", "red"

def estimate_water_loss(flow):
    return round(abs(NORMAL_FLOW - flow), 2)

def generate_live_data():
    zone = np.random.choice(zones)

    flow = np.random.normal(100, 5)
    pressure = np.random.normal(55, 3)
    acoustic = np.random.normal(25, 2)

    df = pd.DataFrame([[flow, pressure, acoustic]],
                      columns=["flow", "pressure", "acoustic"])

    return df, zone, flow, pressure, acoustic

# ---------------- MAIN ----------------
if st.session_state.run:

    data, zone, flow, pressure, acoustic = generate_live_data()

    # FEATURE ENGINEERING
    data["flow_ma"] = data["flow"]
    data["pressure_drop_rate"] = 0
    data["acoustic_change"] = 0
    data["flow_pressure_ratio"] = data["flow"] / data["pressure"]

    data[zone_map_model[zone]] = 1

    for col in model_features:
        if col not in data:
            data[col] = 0

    data = data[model_features]

    scaled = scaler.transform(data)
    prediction = model.predict(scaled)

    st.session_state.flow.append(flow)
    st.session_state.pressure.append(pressure)
    st.session_state.acoustic.append(acoustic)

    # ---------------- STATUS CARDS ----------------
    st.markdown("### 📊 Live System Metrics")

    m1, m2, m3 = st.columns(3)

    m1.metric("💧 Flow", f"{flow:.2f}")
    m2.metric("🔵 Pressure", f"{pressure:.2f}")
    m3.metric("🔊 Acoustic", f"{acoustic:.2f}")

    st.info(f"📍 Monitoring Zone: {zone}")

    # ---------------- LAYOUT ----------------
    left, right = st.columns([2, 1])

    # ================= CHARTS =================
    with left:

        chart_data = pd.DataFrame({
            "Flow": st.session_state.flow,
            "Pressure": st.session_state.pressure,
            "Acoustic": st.session_state.acoustic
        })

        c1, c2, c3 = st.columns(3)

        c1.plotly_chart(px.line(chart_data, y="Flow", title="Flow Trend"), use_container_width=True)
        c2.plotly_chart(px.line(chart_data, y="Pressure", title="Pressure Trend"), use_container_width=True)
        c3.plotly_chart(px.line(chart_data, y="Acoustic", title="Acoustic Trend"), use_container_width=True)

    # ================= LEAK DETECTION =================
    leak_detected = prediction[0] == -1

    if leak_detected:

        severity, color = get_severity(flow, pressure)
        water_loss = estimate_water_loss(flow)

        with left:
            st.markdown(f"## 🚨 Leak Detected — {severity}")
            st.markdown(f"### 💧 Water Loss: {water_loss} L/min")

        with right:

            st.subheader("🗺 Leak Location")

            m = folium.Map(location=zone_coordinates[zone], zoom_start=14)

            folium.Marker(
                location=zone_coordinates[zone],
                popup=f"Leak in {zone}",
                icon=folium.Icon(color=color),
            ).add_to(m)

            st.session_state.map_key += 1

            st_folium(m, width=450, height=450, key=f"map_{st.session_state.map_key}")

    else:
        with left:
            st.markdown("## ✅ System Operating Normally")

        with right:
            st.info("🗺 Map will appear when a leak is detected")

    time.sleep(3)
    st.rerun()

else:
    st.warning("⏹ Monitoring stopped. Click Start to begin.")
```
